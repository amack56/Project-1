```python

import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
import warnings
```

```python
data=pd.read_csv("./../Data/CleanData/full_data.csv")
```

```python
data_2019=data[data["Year"]==2019]
```

```python
player_grouped_data=data_2019.groupby("Player").mean()
```

```python
average_score=player_grouped_data[player_grouped_data["FantasyPoints"]>(15*16)].sort_values("FantasyPoints")
```

```python
fig=plt.figure(figsize=(18,6))
plt.bar(average_score.index, average_score["FantasyPoints"], color ='maroon',
        width = 0.4)
 
plt.xlabel("Players",size=14)
plt.ylabel("Average Score",size=14)
plt.title("Player averaging over 15 points per game for the year 2019",size=15)
plt.xticks(rotation=90,size=13)
plt.show()
```

```python
avg_points_per_pos = data_2019.groupby('Pos').mean()
```

```python
avg_points_per_pos=avg_points_per_pos.iloc[1:,]
```

```python
plt.figure(figsize=(10,7))
plt.bar(avg_points_per_pos.index,avg_points_per_pos["FantasyPoints"])
plt.title('Average Number of points by position for year 2019',size=15)
plt.xlabel('Position',size=14)
plt.ylabel('Average number of points',size=14)
plt.show()
```
